```python
import numpy as np
#np.seterr(divide='ignore', invalid='ignore')
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

sflg='1.erg'
clmn=3
#
a=pd.read_csv(sflg,sep='\s+')
mtx=a.values.T
#print(mtx)
y=mtx[clmn:clmn+1,]
y_1=y.reshape(1,-1)
#y=np.nan_to_num(y_1[0])
y=y_1[0]
x1=0;x2=2 # scale the x plot range


plt.hist(y,bins=200,normed=True,range=(x1,x2),histtype='step')
y1,y2=plt.ylim()
plt.ylabel('Normalized probability density') # 给 x 轴添加标签
plt.xlabel('Potential Energy [a.u.]') # 给 y 轴添加标签

y=y.mean()

u_real=0.25/np.tanh(0.5*8)
#plt.axvline(x=u_real,color='g',ls='-')
plt.axvline(x=y,color='r',ls='--')
plt.title(r'system: quartic potential of $U=\frac{1}{4}x^4$')
plt.text(0.7*x2,0.7*y2,r'$\beta$ = 8, m = 1'+'\n'+r'$\gamma=1$'+'\ndt=0.01 \nntraj=150 \nbeads=200')
plt.text(0.7*x2,0.5*y2,r'$u_{md}=$%f'%y)
#plt.text(0.5*x2,0.5*y2,r'$u_{md}=$%f'%y+'\n'+r'$u_{th}=$%f'%u_real)
print(y)
plt.savefig(sflg+'%d'%clmn+'.png')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#
dt=0.01
ss='0319x/P200j50_b8_0.01_Qt/cx_2.gp'
#
a=pd.read_csv(ss,sep='\s+',header=None)
mtx=a.values.T
x=mtx[0,]/1000
y=mtx[1:,]

rangcct=1000
u0ui=np.zeros((rangcct,len(y)))
I1=np.ones(rangcct)
for i in range(1,rangcct+1):
    z1=y.T[:-i,]
    z2=y.T[i:,]
    cov=np.multiply(z1,z2)
    u0ui[i-1]=cov.mean(axis=0)
um=np.outer(I1,y.mean(axis=1))
um2=np.multiply(um,um)
yy=np.multiply(y,y)
u2m=np.outer(I1,yy.mean(axis=1))
cct=(u0ui-um2)/(u2m-um2)
#
dn=[i*dt for i in range(0,rangcct)]
#
scl=600 # adjust the plot range
mcct=cct.T.mean(axis=0)
#
plt.plot(dn[:scl],mcct[:scl],'r--')
rslt=cct.T.mean(axis=0)
rslt=rslt[:30]
rslt=rslt.sum()*dt
print(rslt)
plt.ylabel('Character Corelation Function of\n Potential of Standard') # 给 x 轴添加标签
plt.xlabel('time [a.u.]') # 给 y 轴添加标签

plt.savefig('cct_x.png')
plt.show()

lst=pd.DataFrame({'cct_x':mcct})
print(lst)
lst.to_csv('save_x.dat')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

val=pd.read_csv('save_x.dat')
t=val.values.T[0]
cct=val.values.T[1]

N=len(cct)

T=len(t)
cut=(N+1)//2
cut1=N-cut

cw = np.fft.fft(cct)

f=np.arange(N)/T
tmp=(1-np.exp(-8*f[0:cut]))/(8*f[0:cut])*cw[0:cut]
tmp[0]=cw[0]
#print(cw)
#
kw=np.arange(N)
kw[0:cut]=tmp.copy()
kw[cut1:N]=tmp[::-1]

#exit()

kcct=np.fft.ifft(kw)

plt.plot(t,cct,'r-',t[0:cut],kcct[0:cut],'b--')
plt.show()
```
